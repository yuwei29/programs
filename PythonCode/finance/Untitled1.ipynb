```python
def getTop500(file):
    x = pd.read_csv(file,usecols=[1],nrows=500)
    x = x['收盘'].tolist()
    x= [float(e.replace(',','')) for e in x]
    x = np.array(x)
    return x
```

```python
nasdaq = getTop500('纳斯达克综合指数历史数据2.csv')
```

```python
vni = getTop500('越南胡志明指数历史数据.csv')
```

```python
sh = getTop500('上证指数历史数据.csv')
```

```python
x=np.arange(500)
y1 = nasdaq - 3*vni
y2 = sh
plt.plot(x,y1,lable='nasdaq-3vni')
plt.plot(x,y2,label='sh')
```

```python
x=np.arange(500)
y1 = nasdaq - 3*vni
y2 = sh
plt.plot(x,y1,label='nasdaq-3vni')
plt.plot(x,y2,label='sh')
```

```python
x=np.arange(8)
y1=4*x+5
y2=3*x+5
y3=2*x+5
y4=x+5

plt.plot(x,y1,label="4x+5")
plt.plot(x,y2,label="3x+5")
plt.plot(x,y3,label="2x+5")
plt.plot(x,y4,label="x+5")

plt.title("Plot Mutiple lines in Matplotlib",fontsize=15)
plt.xlabel("X",fontsize=13)
plt.ylabel("Y",fontsize=13)
plt.legend()
plt.show()
```
